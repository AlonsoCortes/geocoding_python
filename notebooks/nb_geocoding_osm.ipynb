```python
import os
import pandas as pd
import numpy as np
import requests
import time
pd.set_option('display.max_columns', None)
```

```python
# nombre del archivo con la información a geocodificar y su terminación (CSV)
archivo = "reniecyt_2.csv"
```

```python
# Lectura
df = pd.read_csv(os.path.join('datos',archivo ))
print(df.shape)
df.head(1)
```

```python
# Especificar la columna que tiene la ruta/nombre a geocodificar
columna = 'direccion' # Modificar en caso de ser necesario
```

```python
# Configuración de la API de Nominatim
NOMINATIM_URL = "https://nominatim.openstreetmap.org/search"
HEADERS = {"User-Agent": "MiScriptGeocodificador/1.0"}
```

```python
# Función para buscar un lugar y obtener sus coordenadas
def buscar_lugar(lugar):
    try:
        params = {"q": lugar, "format": "json", "limit": 1}  # Busca solo el mejor resultado
        response = requests.get(NOMINATIM_URL, params=params, headers=HEADERS, timeout=10)
        
        if response.status_code == 200:
            data = response.json()
            if data:
                return data[0]["lat"], data[0]["lon"]  # Devuelve latitud y longitud
    except Exception as e:
        print(f"Error con '{lugar}': {e}")

    return None, None  # Si hay error, devuelve valores vacíos
```

```python
# Aplicar la función al DataFrame
df["latitud"], df["longitud"] = zip(*df[columna].apply(buscar_lugar))
```

```python
# Función para obtener coordenadas con Nominatim
def obtener_coordenadas_osm(direccion):
    url = f"https://nominatim.openstreetmap.org/search?q={direccion}&format=json&limit=1"
    headers = {"User-Agent": "MiAplicacion/1.0"}  # Requerido para evitar bloqueos
    try:
        respuesta = requests.get(url, headers=headers)
        datos = respuesta.json()
        if datos:
            return datos[0]["lat"], datos[0]["lon"]
    except Exception as e:
        print(f"Error con '{direccion}': {e}")
    return None, None
```

```python
# Aplicar la función y esperar para no exceder el límite de consultas
df["lat"], df["long"] = zip(*df[columna].apply(lambda x: (time.sleep(1), obtener_coordenadas_osm(x))[1]))
```
